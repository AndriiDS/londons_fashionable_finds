- **Rating**: The rating of a shop indicates its overall customer satisfaction and quality of service. Higher ratings are generally associated with better shopping experiences and can influence the perceived value of the shop. Including rating as a factor in the composite score helps identify shops with higher customer satisfaction.
- **Property_Value_2017**: The property value of the shop's location in 2017 can be an important factor when assessing the shop's worth. A higher property value may indicate a more desirable location, which could have a positive impact on the shop's performance. Including property value as a factor helps capture the influence of location on the shop's overall value.
- **Earnings_2021**: The shop's earnings in 2021 reflect its financial performance and profitability. Higher earnings suggest a more successful and potentially more valuable shop. Including earnings as a factor in the composite score helps capture the shop's financial strength and value.
- **Reviews_Count**: The number of reviews a shop has received can be considered a proxy for the shop's popularity. More reviews often indicate a higher level of customer engagement and interest in the shop. Including the review count as a factor in the composite score helps capture the influence of a shop's popularity on its overall value.
- **Price_Level**: The price level of a shop represents the general price range of its products and services. Higher price levels often indicate a more upscale and exclusive shopping experience. Including the price level as a factor in the composite score helps capture the influence of the shop's pricing strategy on its overall value.


```python
import pandas as pd

# Define the importance weights for each factor
weight_rating = 0.4
weight_reviews_count = 0.3
weight_property_value = 0.2
weight_earnings = 0.2
weight_price_level = 0.1

# Replace missing values in Price_Level column with the mean value
shops['Price_Level'].fillna(shops['Price_Level'].mean(), inplace=True)

# Normalize the variables
shops['Rating_normalized'] = (shops['Rating'] - shops['Rating'].min()) / (shops['Rating'].max() - shops['Rating'].min())
shops['Property_Value_2017_normalized'] = (shops['Property_Value_2017'] - shops['Property_Value_2017'].min()) / (shops['Property_Value_2017'].max() - shops['Property_Value_2017'].min())
shops['Earnings_2021_normalized'] = (shops['Earnings_2021'] - shops['Earnings_2021'].min()) / (shops['Earnings_2021'].max() - shops['Earnings_2021'].min())
shops['Reviews_Count_normalized'] = (shops['Reviews_Count'] - shops['Reviews_Count'].min()) / (shops['Reviews_Count'].max() - shops['Reviews_Count'].min())
shops['Price_Level_normalized'] = (shops['Price_Level'] - shops['Price_Level'].min()) / (shops['Price_Level'].max() - shops['Price_Level'].min())

# Calculate the composite score for each shop using the weighted scoring model and normalized variables
shops['composite_score'] = (shops['Rating_normalized'] * weight_rating) + \
                           (shops['Property_Value_2017_normalized'] * weight_property_value) + \
                           (shops['Earnings_2021_normalized'] * weight_earnings) + \
                           (shops['Reviews_Count_normalized'] * weight_reviews_count) + \
                           (shops['Price_Level_normalized'] * weight_price_level)

# Rank the shops by composite score in descending order
ranked_shops = shops.sort_values('composite_score', ascending=False)
```

```python
# Add a new 'Rank' column to the ranked_shops DataFrame
ranked_shops['Rank'] = ranked_shops['composite_score'].rank(method='min', ascending=False)

# Reset the index to reflect the new ranking order
ranked_shops.reset_index(drop=True, inplace=True)

# Display the ranked shops with their respective rank
display(ranked_shops[['Shop_Name', 'Address', 'composite_score', 'Rank']])
```
