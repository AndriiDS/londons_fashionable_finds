```python
import pandas as pd

# Load the CSV into a DataFrame
shops = pd.read_csv('03_Feature_Engineering/shops.csv')
shops = shops.drop(columns=['Unnamed: 0'])

```

```python
import pandas as pd

# Define the importance weights for each factor
weight_rating = 0.4
weight_property_value = 0.3
weight_earnings = 0.3

# Calculate the composite score for each shop using the weighted scoring model
shops['composite_score'] = (shops['Rating'] * weight_rating) + \
                                   (shops['Property_Value_2017'] * weight_property_value) + \
                                   (shops['Earnings_2021'] * weight_earnings)

# Rank the shops by composite score in descending order
ranked_shops = shops.sort_values('composite_score', ascending=False)

```
